```
import tensorflow as tf
from tensorflow.keras.datasets import fashion_mnist
from tensorflow.keras import models
from tensorflow.keras import layers
import matplotlib.pyplot as plt
```

```
(train_images, train_labels), (test_images, test_labels) = fashion_mnist.load_data()
```

```
# Normalizing the data so we get the pixel values in between 0 and 1
train_images = train_images.astype('float') / 255
test_images = test_images.astype('float') / 255
```

```
# Network Architecture
network = models.Sequential([layers.Flatten(), layers.Dense(512, activation='relu'), layers.Dense(10, activation='softmax')])
```

```
# Compile Network
network.compile(optimizer=tf.keras.optimizers.Adam(), loss='sparse_categorical_crossentropy', metrics=['accuracy'])
```

```
# Train Network
network.fit(train_images, train_labels, epochs=10, batch_size=128)
```

```
loss, acc = network.evaluate(test_images, test_labels)
```
